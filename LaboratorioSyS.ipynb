En este taller construiremos un filtro muy simple que permitirá detectar los bordes en una imagen.
Primero cargaremos las librerías de Python necesarias: matplotlib, numpy y scipy. 

También específicamos la ubicación de dos imágenes que usaremos en este laboratorio. La primera se llama ```elwood.png``` y contiene una imagen de un perro. La otra es ```noisy_elwood.png``` y es la misma imagen anterior que se le ha agregado ruido.

````python
import matplotlib.pyplot as plt
%matplotlib inline


import numpy as np
import scipy.signal as sp

url_elwood = 'https://raw.githubusercontent.com/rodrigocadiz/IEE1100/master/elwood.png'
url_noisy_elwood = 'https://raw.githubusercontent.com/rodrigocadiz/IEE1100/master/noisy_elwood.png'
````

Ahora leeremos el archivo elwood.png, que es una fotografía de *Elwood*, un perro muy famoso en la comunidad de procesamiento de imágenes. Esto se hace con el comando ```imread``` de matplotlib. Este comando lee la imagen del disco y a convierte a una matriz. Podemos imprimir las dimensiones de la matriz, y vemos que tiene 765 píxeles de alto y 1366 de ancho.

````python
elwood = plt.imread(url_elwood)
print(elwood.shape)
````

Los datos de la imagen, es decir el valor de intensidad de cada uno de los píxeles, se encuentran en la matriz llamada elwood. Podemos imprimir todos estos valores con el siguiente comando:

````python
print(elwood)
````

Podemos selecionar un pixel en particular, por ejemplo en la posición 200, 1000 que corresponde a la fila 200 y columna 1000 de la siguiente manera:

````python
pixel = elwood[200,1000]
print(pixel)
````

Podemos mostrar en pantalla la imagen con los siguientes dos comandos de ```matplotlib```.

````python
plt.imshow(elwood, cmap='gray')
plt.show()
````
